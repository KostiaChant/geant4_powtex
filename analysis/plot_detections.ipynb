# POWTEX 3D Event Visualization

This notebook visualizes the 3D positions of detected events from the POWTEX biosample detector, showing both actual detection positions and voxel coordinates.

## 1. Import Required Libraries

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import warnings
warnings.filterwarnings('ignore')
```

## 2. Load and Explore the Data

```python
# Load the CSV file
df = pd.read_csv('../data/output/detections.csv', sep='\t')

print(f"Data shape: {df.shape}")
print(f"\nFirst few rows:")
print(df.head())
print(f"\nData types:")
print(df.dtypes)
print(f"\nBasic statistics:")
print(df[['x_pos [mm]', 'y_pos [mm]', 'z_pos [mm]', 'voxel_x [mm]', 'voxel_y [mm]', 'voxel_z [mm]']].describe())
```

## 3. Prepare Data for 3D Visualization

```python
# Extract coordinate columns
x_pos = df['x_pos [mm]'].values
y_pos = df['y_pos [mm]'].values
z_pos = df['z_pos [mm]'].values

voxel_x = df['voxel_x [mm]'].values
voxel_y = df['voxel_y [mm]'].values
voxel_z = df['voxel_z [mm]'].values

# Check for missing values
print(f"Missing values in actual positions:")
print(f"  x_pos: {df['x_pos [mm]'].isna().sum()}")
print(f"  y_pos: {df['y_pos [mm]'].isna().sum()}")
print(f"  z_pos: {df['z_pos [mm]'].isna().sum()}")

print(f"\nMissing values in voxel coordinates:")
print(f"  voxel_x: {df['voxel_x [mm]'].isna().sum()}")
print(f"  voxel_y: {df['voxel_y [mm]'].isna().sum()}")
print(f"  voxel_z: {df['voxel_z [mm]'].isna().sum()}")

print(f"\nTotal number of events: {len(df)}")
```

## 4. Create 3D Plot of Actual Positions

```python
# Create 3D scatter plot for actual positions
fig_pos = go.Figure(data=[go.Scatter3d(
    x=x_pos,
    y=y_pos,
    z=z_pos,
    mode='markers',
    marker=dict(
        size=4,
        color='blue',
        opacity=0.7,
        line=dict(width=0)
    ),
    text=[f"Event {i}<br>x: {x_pos[i]:.2f}<br>y: {y_pos[i]:.2f}<br>z: {z_pos[i]:.2f}" 
          for i in range(len(x_pos))],
    hovertemplate='%{text}<extra></extra>',
    name='Actual Position'
)])

fig_pos.update_layout(
    title='POWTEX Detector: Actual Detection Positions',
    scene=dict(
        xaxis_title='X Position [mm]',
        yaxis_title='Y Position [mm]',
        zaxis_title='Z Position [mm]',
        aspectmode='data'
    ),
    width=900,
    height=800,
    hovermode='closest'
)

fig_pos.show()
```

## 5. Create 3D Plot of Voxel Coordinates

```python
# Create 3D scatter plot for voxel coordinates
fig_voxel = go.Figure(data=[go.Scatter3d(
    x=voxel_x,
    y=voxel_y,
    z=voxel_z,
    mode='markers',
    marker=dict(
        size=4,
        color='red',
        opacity=0.7,
        line=dict(width=0)
    ),
    text=[f"Event {i}<br>voxel_x: {voxel_x[i]:.2f}<br>voxel_y: {voxel_y[i]:.2f}<br>voxel_z: {voxel_z[i]:.2f}" 
          for i in range(len(voxel_x))],
    hovertemplate='%{text}<extra></extra>',
    name='Voxel Coordinate'
)])

fig_voxel.update_layout(
    title='POWTEX Detector: Voxel Coordinates',
    scene=dict(
        xaxis_title='Voxel X [mm]',
        yaxis_title='Voxel Y [mm]',
        zaxis_title='Voxel Z [mm]',
        aspectmode='data'
    ),
    width=900,
    height=800,
    hovermode='closest'
)

fig_voxel.show()
```

## 6. Create Combined 3D Plot with Color Differentiation

```python
# Create combined 3D scatter plot with both datasets
fig_combined = go.Figure()

# Add actual positions (blue)
fig_combined.add_trace(go.Scatter3d(
    x=x_pos,
    y=y_pos,
    z=z_pos,
    mode='markers',
    marker=dict(
        size=4,
        color='blue',
        opacity=0.6,
        line=dict(width=0)
    ),
    text=[f"Actual Position {i}<br>x: {x_pos[i]:.2f}<br>y: {y_pos[i]:.2f}<br>z: {z_pos[i]:.2f}" 
          for i in range(len(x_pos))],
    hovertemplate='%{text}<extra></extra>',
    name='Actual Detection Position'
))

# Add voxel coordinates (red)
fig_combined.add_trace(go.Scatter3d(
    x=voxel_x,
    y=voxel_y,
    z=voxel_z,
    mode='markers',
    marker=dict(
        size=4,
        color='red',
        opacity=0.6,
        line=dict(width=0)
    ),
    text=[f"Voxel {i}<br>voxel_x: {voxel_x[i]:.2f}<br>voxel_y: {voxel_y[i]:.2f}<br>voxel_z: {voxel_z[i]:.2f}" 
          for i in range(len(voxel_x))],
    hovertemplate='%{text}<extra></extra>',
    name='Voxel Coordinate'
))

fig_combined.update_layout(
    title='POWTEX Detector: Actual Positions vs Voxel Coordinates',
    scene=dict(
        xaxis_title='X / Voxel_X [mm]',
        yaxis_title='Y / Voxel_Y [mm]',
        zaxis_title='Z / Voxel_Z [mm]',
        aspectmode='data'
    ),
    width=1000,
    height=800,
    hovermode='closest',
    showlegend=True,
    legend=dict(
        x=0.7,
        y=0.9,
        bgcolor='rgba(255,255,255,0.8)',
        bordercolor='black',
        borderwidth=1
    )
)

fig_combined.show()
```

## 7. Analysis: Comparison of Actual Positions vs Voxel Coordinates

```python
# Calculate differences between actual positions and voxel coordinates
dx = x_pos - voxel_x
dy = y_pos - voxel_y
dz = z_pos - voxel_z

# Calculate distances
distance_diff = np.sqrt(dx**2 + dy**2 + dz**2)

print("=" * 60)
print("POSITION COMPARISON ANALYSIS")
print("=" * 60)
print(f"\nActual Position Range:")
print(f"  X: [{x_pos.min():.2f}, {x_pos.max():.2f}] mm")
print(f"  Y: [{y_pos.min():.2f}, {y_pos.max():.2f}] mm")
print(f"  Z: [{z_pos.min():.2f}, {z_pos.max():.2f}] mm")

print(f"\nVoxel Coordinate Range:")
print(f"  X: [{voxel_x.min():.2f}, {voxel_x.max():.2f}] mm")
print(f"  Y: [{voxel_y.min():.2f}, {voxel_y.max():.2f}] mm")
print(f"  Z: [{voxel_z.min():.2f}, {voxel_z.max():.2f}] mm")

print(f"\nDifference Statistics (Actual - Voxel):")
print(f"  ΔX: mean={dx.mean():.4f}, std={dx.std():.4f}, min={dx.min():.4f}, max={dx.max():.4f} mm")
print(f"  ΔY: mean={dy.mean():.4f}, std={dy.std():.4f}, min={dy.min():.4f}, max={dy.max():.4f} mm")
print(f"  ΔZ: mean={dz.mean():.4f}, std={dz.std():.4f}, min={dz.min():.4f}, max={dz.max():.4f} mm")

print(f"\nDistances Between Actual and Voxel Coordinates:")
print(f"  Mean distance: {distance_diff.mean():.4f} mm")
print(f"  Std deviation: {distance_diff.std():.4f} mm")
print(f"  Min distance: {distance_diff.min():.4f} mm")
print(f"  Max distance: {distance_diff.max():.4f} mm")
print(f"  Median distance: {np.median(distance_diff):.4f} mm")
```
